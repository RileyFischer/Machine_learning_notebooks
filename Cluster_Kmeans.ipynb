```python
# Load Data
import numpy as np
from google.colab import drive

drive.mount('/content/drive')
X = [0,0,0,0]
with open('/content/drive/My Drive/Colab Notebooks/data/data_uniform.csv', 'r') as f: 
    X[0] = np.genfromtxt(f, dtype='f4', delimiter=',')
with open('/content/drive/My Drive/Colab Notebooks/data/data_noisycircle.csv', 'r') as f: 
    X[1] = np.genfromtxt(f, dtype='f4', delimiter=',')
with open('/content/drive/My Drive/Colab Notebooks/data/data_noisymoon.csv', 'r') as f: 
    X[2] = np.genfromtxt(f, dtype='f4', delimiter=',')
with open('/content/drive/My Drive/Colab Notebooks/data/data_3blobs.csv', 'r') as f: 
    X[3] = np.genfromtxt(f, dtype='f4', delimiter=',')
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

c = 3
f = plt.figure(figsize=(15,15))
for i in range(4):
  kmeans = KMeans(n_clusters=c, random_state=0).fit(X[i])
  ax = f.add_subplot(2,2,i+1)
  scatter = ax.scatter(X[i][:,0], X[i][:,1], c=kmeans.labels_)
  ax.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], c='r', marker='x')
  plt.legend(*scatter.legend_elements())
```
