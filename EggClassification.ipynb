**Egg classification**

classifying eggs (medium, large, jumbo) based on length L and weight W

Label set: 0 (medium), 1 (large), 2 (jumbo)

feature: L: length 0 < L < 10, W: weight 0 < W < 30

implement a rule-based classifier and classify randomly generated feature
vectors (L, W) into one of the three labels

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

N = 1000; # number of randomly generated features
x = np.random.rand(N,2)*[10,30] # Nx2 feature vectors [L, W]

# classification rules:
# if W< 10, L < 3, then label = 0 (medium)  y0=1
# if W> 20, L > 5, then label = 2 (jumbo)   y2=1
# else label = 1 (large)   y1 = 1

# indices of Medium eggs
id0 = np.logical_and(x[:,0]<3,x[:,1]<10)
# indices of Large eggs
id1 = np.logical_and(x[:,0]>5,x[:,1]>20)
# indices of Jumble eggs
id2 = np.logical_not(np.logical_or(id0,id1))
```

```python
plt.xlabel('Length')
plt.ylabel('Weight')
plt.title('classification results')
plt.plot(x[id0,0],x[id0,1],'ob',x[id1,0],x[id1,1],'sg',x[id2,0],x[id2,1],'xr')
plt.legend(('medium','large','jumbo'),loc='lower right')
plt.show()
```
