# Perceptron Example
Perceptron learning demonstration

Given a set of linearly separable training data, use the perceptron learning algorithm to separate these two classes of data. We use the iris dataset, features 3 and 4. 

**1) Load the iris dataset**

```python
import numpy as np
from google.colab import drive

drive.mount('/content/drive')
with open('/content/drive/My Drive/Colab Notebooks/data/iris.csv', 'r') as f: 
  temp = np.genfromtxt(f, dtype='f4', delimiter=',')

X = temp[:,2:4]
target = temp[:,4] == 1
c1 = np.where(target==1)
c2 = np.where(target==0)
```

Scatter plot the 2D data

```python
import matplotlib.pyplot as plt
ax = [0,8,0,3]
plt.scatter(X[c1,0],X[c1,1],c='white', edgecolors='green')
plt.scatter(X[c2,0],X[c2,1],c='white',edgecolors='magenta')
plt.xlabel('3rd feature')
plt.ylabel('4th Feature')
plt.title('3rd and 4th dimension of the Iris dataset')
plt.xlim((ax[0],ax[1]))
plt.ylim((ax[2],ax[3]))
plt.show()
```

**2) Call sklearn perceptron package to evaluate linear separable hyperplane.**


```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,  y_test = train_test_split(X, target, test_size=0.2)
```

```python
from sklearn.linear_model import Perceptron
pModel = Perceptron(random_state=42).fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

predictions_test = pModel.predict(X_test)
test_score = accuracy_score(predictions_test, y_test)
print("score on test data: ", test_score)
```
