**Apply PCA to iris using 2 principal components**

```python
import numpy as np
from google.colab import drive

drive.mount('/content/drive')
with open('/content/drive/My Drive/Colab Notebooks/data/iris.csv', 'r') as f: 
  A = np.genfromtxt(f,delimiter=',')
X = A[:,:-1]
y = A[:,-1]
```

```python
from sklearn.decomposition import PCA

pcaModel = PCA(n_components=2)
X_r = pcaModel.fit(X).transform(X)
```

```python
import matplotlib.pyplot as plt

# Scatter plot the 2D data 

plt.scatter(X_r[:,0],X_r[:,1],c=y)
plt.xlabel('1st PC')
plt.ylabel('2nd PC')
plt.title('Iris data using 2 PCs')
plt.show()
```
