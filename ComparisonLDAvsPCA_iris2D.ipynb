```python
import numpy as np
from google.colab import drive

drive.mount('/content/drive')
with open('/content/drive/My Drive/Colab Notebooks/data/iris.csv', 'r') as f: 
  A = np.genfromtxt(f,delimiter=',')
X = A[:,:-1]
y = A[:,-1]
```

```python
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
import matplotlib.pyplot as plt
#https://scikit-learn.org/stable/auto_examples/decomposition/plot_pca_vs_lda.html

pca = PCA(n_components=2)
X_r = pca.fit(X).transform(X)
# Percentage of variance explained for each components
print('explained variance ratio (first two components): %s'
      % str(pca.explained_variance_ratio_))
```

```python
lda = LinearDiscriminantAnalysis(n_components=2)
X_r2 = lda.fit(X, y).transform(X)
```

```python
plt.figure()
colors = ['navy', 'turquoise', 'darkorange']

for color, i in zip(colors, [1,2,3]):
    plt.scatter(X_r[y == i, 0],X_r[y == i, 1],color=color, alpha=.8)
plt.title('PCA of IRIS dataset')

# another way to graph diff color for diff label
plt.figure()
plt.scatter(X_r2[:,0],X_r2[:,1],c=y,alpha=.8)
plt.title('LDA of IRIS dataset')

plt.show()
```
